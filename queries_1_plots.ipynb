```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def set_plot_size():
    figure = plt.gcf()
    px = 1 / 141
    figure.set_size_inches(1920 * px, 1080 * px)
```

```python
query_folder_path = 'queries_1/movie_genres_sorted'
query_file_path = f'{query_folder_path}/part-r-00000'
df = pd.read_csv(query_file_path, sep='\t', names=['genre', 'count'])
print(df)
```

```python
plt.bar(df['genre'], df['count'], color=['#d42e12', '#666666'])
plt.title(f'Quantity of movies by genre')

plt.xticks(rotation=90)
plt.ylim(0, 2_000)

plt.grid(axis='y', linestyle='--')
set_plot_size()

figure_path = f'{query_folder_path}/movie_genres_count.png'
plt.savefig(figure_path, dpi=460, bbox_inches='tight')
plt.show()
```

```python
query_folder_path = 'queries_1/movie_year_genres_sorted'
query_file_path = f'{query_folder_path}/part-r-00000'
df = pd.read_csv(query_file_path, sep='\t', names=['year', 'genre', 'count'])
print(df)
```

```python
filtered_df = df[df['count'] >= 25]
print(filtered_df)
```

```python
grouped_df = pd.DataFrame()
grouped_df['year_genre'] = filtered_df['year'].map(str) + ', ' + filtered_df['genre']
grouped_df['count'] = filtered_df['count']
print(grouped_df)
```

```python
plt.bar(grouped_df['year_genre'], grouped_df['count'], color=['#d42e12', '#666666'])
plt.title(f'Quantity of movies by year and genre')

plt.xticks(rotation=90)
plt.ylim(0, 200)

plt.grid(axis='y', linestyle='--')
set_plot_size()

figure_path = f'{query_folder_path}/movie_year_genres_count.png'
plt.savefig(figure_path, dpi=460, bbox_inches='tight')
plt.show()
```

```python
query_folder_path = 'queries_1/movie_year_sorted'
query_file_path = f'{query_folder_path}/part-r-00000'
df = pd.read_csv(query_file_path, sep='\t', names=['year', 'count'])
print(df)
```

```python
query_folder_path = 'queries_1/users_age_sorted'
query_file_path = f'{query_folder_path}/part-r-00000'
df = pd.read_csv(query_file_path, sep='\t', names=['age', 'count'])
print(df)
```

```python
resorted_df = df.sort_values(by=['age'])
print(resorted_df)
```

```python
resorted_df['age_range'] = ['Under 18', '18-24', '25-34', '35-44', '45-49', '50-55', '56+']
print(resorted_df)
```

```python
plt.pie(resorted_df['count'], labels=resorted_df['age_range'], autopct='%1.1f%%')
plt.title(f'Percentage of users by age range')

set_plot_size()

figure_path = f'{query_folder_path}/users_age_percentage.png'
plt.savefig(figure_path, bbox_inches='tight')
plt.show()
```
