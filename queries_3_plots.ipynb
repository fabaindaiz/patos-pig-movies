```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def set_plot_size():
    figure = plt.gcf()
    px = 1 / 141
    figure.set_size_inches(1920 * px, 1080 * px)

def set_percentage_labels():
    percentage_list = [i for i in range(0, 100 + 1, 10)]
    plt.yticks(ticks=percentage_list, labels=percentage_list)
    plt.ylabel(f'Percentage')
    plt.ylim(-5, 105)
```

```python
query_folder_path = 'queries_3/proportion_votes_age'
query_file_path = f'{query_folder_path}/part-r-00000'
df = pd.read_csv(query_file_path, sep='\t', names=['age', 'proportion'])
print(df)
```

```python
percentage_df = pd.DataFrame()
percentage_df['age_range'] = ['Under 18', '18-24', '25-34', '35-44', '45-49', '50-55', '56+']
percentage_df['percentage'] = df['proportion'] * 100
print(percentage_df)
```

```python
plt.pie(percentage_df['percentage'], labels=percentage_df['age_range'], autopct='%1.1f%%')
plt.title(f'Percentage of votes by age range')

set_plot_size()

figure_path = f'{query_folder_path}/votes_age_percentage.png'
plt.savefig(figure_path, bbox_inches='tight')
plt.show()
```

```python
query_folder_path = 'queries_3/proportion_votes_gender_age'
query_file_path = f'{query_folder_path}/part-r-00000'
df = pd.read_csv(query_file_path, sep='\t', names=['gender', 'age', 'proportion'])
print(df)
```

```python
percentage_df = pd.DataFrame()

gender_dict = {'F': 'Female', 'M': 'Male'}
percentage_df['gender'] = df['gender'].map(gender_dict)

age_range_dict = {1: 'Under 18', 18: '18-24', 25: '25-34', 35: '35-44', 45: '45-49', 50: '50-55', 56: '56+'}
percentage_df['age_range'] =  df['age'].map(age_range_dict)

percentage_df['percentage'] = df['proportion'] * 100
print(percentage_df)
```

```python
grouped_df = pd.DataFrame()
grouped_df['gender_age'] = percentage_df['gender'].map(str) + ', ' + percentage_df['age_range']
grouped_df['percentage'] = percentage_df['percentage']
print(grouped_df)
```

```python
plt.pie(grouped_df['percentage'], labels=grouped_df['gender_age'], autopct='%1.1f%%')
plt.title(f'Percentage of votes by gender and age')

set_plot_size()

figure_path = f'{query_folder_path}/votes_gender_age_percentage.png'
plt.savefig(figure_path, bbox_inches='tight')
plt.show()
```

```python
query_folder_path = 'queries_3/sorted_count_votes_movies'
query_file_path = f'{query_folder_path}/part-r-00000'
df = pd.read_csv(query_file_path, sep='\t', names=['title', 'year', 'votes'])
print(df)
```

```python
filtered_df = df[df['votes'] >= 1500]
print(filtered_df)
```

```python
plt.bar(filtered_df['title'], filtered_df['votes'], color=['#d42e12', '#666666'])
plt.title(f'Quantity of votes by movie')

plt.xticks(rotation=90)
plt.ylim(0, 3500)

plt.grid(axis='y', linestyle='--')
set_plot_size()

figure_path = f'{query_folder_path}/movies_votes_count.png'
plt.savefig(figure_path, dpi=460, bbox_inches='tight')
plt.show()
```
